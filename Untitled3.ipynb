```
df=pd.read_csv('서울시 코로나19 확진자 현황 (1).csv', encoding = 'cp949', parse_dates=['확진일','등록일','수정일'])
```

```
df=df.dropna(axis=1)
```

```
date_list = df['확진일'].apply(lambda x: re.sub('[^0-9]','',x))
```

```
index_list =[]

for i in range(len(date_list)):
    if len(date_list[i])<4 :
        index_list.append(i)
```

```
date_list2=[]

for i in index_list:
    a= '0'+str(date_list[i])
    date_list2.append(a)
```

```

date_list[index_list]=date_list2
```

```
date_list.value_counts()
```

```
month_list = []
for i in date_list :
    a=str(i)[:2]
    month_list.append(a)
```

```
df['확진월']=month_list
```

```
df.head()
```

```
day_list = []
for i in date_list :
    if len(i)>3:
        a=str(i)[2:]
        day_list.append(a)
    else:
        b=str(i)[3:]
        day_list.append(b)
```

```
df['확진일']=day_list
```

## ▶월별 확진자 현황 분석 

```
df.groupby('확진월').count()['연번'].plot(kind='line', figsize=(12,7))
```

### 확진 월별 현황을 보니 확진자가 급등한 시기는 7~9월 입니다. 해당월을 상세하게 분석해보도록 하겠습니다. 

## ▶07~09월 확진자 감염경로 

```
top_month = ['07','08','09']

df[df['확진월'].isin(top_month)].groupby('접촉력').count()['연번'].head(20).plot(kind='barh', figsize=(12,7), color='red', alpha=0.5)
plt.title('확진자가 폭등한 07~09월 감염경로')
```

### 확진자 감염경로를 확인해보면 가장 많은 것은 '감염경로 조사중'이며, 그 다음으로 높은 항목은 '8.15 서울도심집회'입니다. 

```
df['확진일']=df['확진일'].replace('','0')
df['확진일']=df['확진일'].astype(int)
df[df['접촉력']=='8.15서울도심집회']['확진일'].value_counts()
```

### 8.15 도심집회 참석자들의 확진일을 살펴보면 20~30일 사이에 가장 많은 확진자가 발생했습니다. 

```
df[df['접촉력']=='8.15서울도심집회']['확진일'].plot(kind='kde')
```

### 집회 참석자들의 지역을 살펴보면 강서구가 가장 많습니다. 

```
df[df['접촉력']=='8.15서울도심집회']['지역'].value_counts().plot(kind='barh', figsize=(12,7), color='red', alpha=0.4)
```

## ▶지역별 발생 현황 

```
df['지역'].value_counts().sort_values(ascending=False).head(20).plot(kind='barh', figsize=(12,7), color='red',alpha=0.5)
```

### 관악구에서 코로나 발생이 가장 많았음을 알 수 있습니다. 관악구에서 가장 많은 확진자를 발생시킨 요인은 '기타 확진자 접촉'입니다. 

```
df[df['지역']=='관악구']['접촉력'].value_counts().head(20).plot(kind='barh', color='red', alpha=0.5, figsize=(12,7))
```
