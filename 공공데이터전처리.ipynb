# 서울, 1인가구 수의 추이를 확인하기 위한 데이터 전처리 

- 행정구역에서 서울만 분리 후 시군구명으로 정렬
- 1~10세대 중 1세대와 전체세대만 분리 
- 전체세대 대비 1세대의 비율 칼럼 추가 


```
df=df.set_index('행정구역')
```

```
single = df.transpose()[df.transpose().index.str.contains('1인세대')]
total = df.transpose()[df.transpose().index.str.contains('전체세대')]
```

```
single.head()
```

```
single = single.transpose()
single.head()
```

```
seoul_single=single[single.index.str.contains('서울특별시')]
seoul_single.head()
```

```
seoul_single=seoul_single.reset_index()
```

```
seoul_single['행정구역']
```

```
resub_list = seoul_single['행정구역'].apply(lambda x: re.sub('[^가-힣]','',x))

drop_list = []

for i in range(len(resub_list)):
    if len(resub_list[i]) <9:
        print(i)
        drop_list.append(i)
```

```
resub_list
```

```
seoul_single_clean = seoul_single.iloc[drop_list].reset_index(drop=True)
```

```
seoul_single_clean=seoul_single_clean.iloc[1:,:]
```

```
seoul_single_clean.head()
```

```
seoul_single_clean.columns =['행정구역','1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']
seoul_single_clean.head()
```

```
seoul_single_clean = seoul_single_clean.reset_index(drop=True)
```

```
region_list = []

for i in seoul_single_clean.index:
   a=seoul_single_clean['행정구역'][i].split(' ')[1]
   region_list.append(a)
```

```
seoul_single_clean['행정구역']=region_list
```

```
seoul_single_clean.head()
```

```
seoul_single_clean.info()
```

```
int_list = ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월',
       '11월', '12월']
```

```
for i in int_list:
    seoul_single_clean[i]=seoul_single_clean[i].apply(lambda x: re.sub(',','',x)).astype(int)
```
